```python
train = df[df['Date']<'2015-02-20']
test = df[df['Date']>='2015-02-20']
```

```python
# Removing punctuations
data=train.iloc[:,2:]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data.head()
```

```python
list = [i for i in range(25)]
new_index = [str(i) for i in list]
data.columns= new_index
data.head()
```

```python
for index in new_index:
    data[index] = data[index].str.lower()
    
data.head(2)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
headlines[1:3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
#Implement RFClassifier
model = RandomForestClassifier(n_estimators=200,criterion='entropy')
model.fit(traindataset,train['Label'])
```

```python
data2=test.iloc[:,2:]
data2.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data2.head()
```

```python
list2 = [i for i in range(25)]
new_index2 = [str(i) for i in list2]
data2.columns= new_index
data2.head(2)
```

```python
for index in new_index2:
    data2[index] = data2[index].str.lower()
    
data2.head(2)
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(data2.index)):
    test_transform.append(' '.join(str(x) for x in data2.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = model.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
